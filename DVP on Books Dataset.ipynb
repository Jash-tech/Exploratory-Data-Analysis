# Introduction to Dataset


Dataset on Amazon's Top 50 bestselling books from 2009 to 2019. Contains 550 books,
data has been categorized into fiction and non-fiction using Good reads.
About Dataset:-
1. Name: Name of the Book
2. Author: The author of the Book
3. User Rating: Amazon User Rating
4. Reviews: Number of written reviews on amazon
5. Price: The price of the book (As of 13/10/2020)
6. Year: The Year(s) it ranked on the bestseller
7. Genre: Whether fiction or non-fiction

# Amount of Type of Books Sold on Amazon
Over the 10 years, we see there are two genre of books that are Fiction and Non-Fiction.<br>
We use Seaborn Countplot to display the count of Non- Fiction and Fiction being sold on Amazon.<br>
We see the number of Non - Fiction books are sold more than Fiction books.

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Genre',data=df)
plt.show()
```

# Trend of Ratings over the 10 years.<br>
We use seaborn lineplot to draw relationship between ratings and years feature in the dataset.<br>
There is an decreasing trend for user ratings on Amazon from 2009 to 2012. While from 2012 to 2019 we see an upward trend which indicates that a greater number of users started  giving good feedback on the books purchased from 2012 to 2019. 

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Year',y='User Rating',data=df,ci=None)

plt.show()
```

# Trend of Ratings over the Genre for Years.<br>
We use seaborn lineplot to draw relationship between ratings and years feature in the dataset, taking here consideration for the genre of books sold.<br>
We have two genre of books in dataset that is Fiction and Non-Fiction.<br>
We see for Fiction type, there is steep decrease in ratings for year 2011 to2012.After that there is increase in the trend of ratings from year 2012 to 2019, which means that people started giving good feedback for Fiction books over Non Fiction ones in the Amazon. 

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Year',y='User Rating',hue='Genre',data=df,ci=None)

plt.show()
```

# Year wise count of books with particular rating
We draw bar plot with .plot() method to show the counts of books with different rating in a particular year.
After that we use subplot() method to draw multiple graphs in a single frame for better understanding.<br>
Below plot shows that number of books in a datset with ratings 4.9, 4.5 and 4.4 is significantly less over the years as compared to those books with ratings 4.8, 4.7 and 4.6 

```python

figure, axis = plt.subplots(3, 2,figsize=(11,13))
figure.tight_layout(pad=5.0)
  

df['Year'].loc[df['User Rating']==4.9].value_counts().sort_index().plot(kind='bar',ax=axis[0, 0],color='y')

axis[0, 0].set_title("Yearwise count of books with rating 4.9")
  

df['Year'].loc[df['User Rating']==4.8].value_counts().sort_index().plot(kind='bar',ax=axis[0, 1],color='c')
axis[0, 1].set_title("Yearwise count of books with rating 4.8")
  

df['Year'].loc[df['User Rating']==4.7].value_counts().sort_index().plot(kind='bar',ax=axis[1, 0],color='g')
axis[1, 0].set_title("Yearwise count of books with rating 4.7")
  

df['Year'].loc[df['User Rating']==4.5].value_counts().sort_index().plot(kind='bar',ax=axis[1, 1],color='r')
axis[1, 1].set_title("Yearwise count of books with rating 4.5")

df['Year'].loc[df['User Rating']==4.6].value_counts().sort_index().plot(kind='bar',ax=axis[2, 0],color='b')
axis[2, 0].set_title("Yearwise count of books with rating 4.6")

df['Year'].loc[df['User Rating']==4.4].value_counts().sort_index().plot(kind='bar',ax=axis[2, 1],color='m')
axis[2, 1].set_title("Yearwise count of books with rating 4.4")

axis[0,0].set(xlabel="Years",ylabel="Counts")
axis[1,0].set(xlabel="Years",ylabel="Counts")
axis[0,1].set(xlabel="Years",ylabel="Counts")
axis[1,1].set(xlabel="Years",ylabel="Counts")
axis[2,0].set(xlabel="Years",ylabel="Counts")
axis[2,1].set(xlabel="Years",ylabel="Counts")

plt.show()
```
